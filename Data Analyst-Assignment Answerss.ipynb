Question 2:
Drop the columns "id" and "Unnamed: 0" from axis 1 using the method drop(), then use

submit it, make sure the inplace parameter is set to True.

containing NaN values. Replace the Nan values with the mean value of that column.

```python
# Drop the columns 'id' and 'Unnamed: 0'
data.drop(['id', 'Unnamed: 0'], axis=1, inplace=True, errors='ignore')

# Get a statistical summary of the data
description = data.describe()
print(description)

# Check for NaN values and print the count for each column containing NaNs
nan_counts = data.isna().sum()
nan_columns = nan_counts[nan_counts > 0]
print('Columns with NaN values and their counts:')
print(nan_columns)

# Replace NaN values with the mean of their respective columns
data.fillna(data.mean(), inplace=True)

# Verify that NaN values have been replaced
nan_counts_after = data.isna().sum()
print('NaN values after replacement:')
print(nan_counts_after)
```

The code executed successfully for dropping the specified columns and generating a statistical summary. However, there was an issue when attempting to replace  values with the mean of their respective columns, likely due to the presence of non-numeric data in some columns.

```python
# Check the data types of each column to identify non-numeric columns
column_types = data.dtypes
print(column_types)
```

```python
# Identify numeric columns
numeric_columns = data.select_dtypes(include=['number']).columns

# Print the numeric columns
print('Numeric columns:')
print(numeric_columns)

# Replace NaN values with the mean of their respective numeric columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())

# Verify that NaN values have been replaced
nan_counts_after = data.isna().sum()
print('NaN values after replacement:')
print(nan_counts_after[nan_counts_after > 0])
```

Question 6
Fit a linear regression model to predict the 'price' using the feature 'sqft_living' then calculate
the R^2. Take a screenshot of your code and the value of the R^2.

Answer:The linear regression model has been successfully fitted using the sqft_living feature to predict house prices, and the calculated R value is approximately 0.494. This indicates that about 49.4% of the variance in house prices can be explained by the sqft_living feature.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Prepare the data for modeling
X = data[['sqft_living']]
Y = data['price']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)

# Make predictions and calculate R^2
Y_pred = model.predict(X_test)
r2 = r2_score(Y_test, Y_pred)

# Print the R^2 value
print('R^2 value:', r2)
```

Question 7
Fit a linear regression model to predict the 'price' using the list of features:
["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view"
,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]


Answer:The linear regression model has been successfully fitted using the specified features to predict the 'price', and the calculated R value is approximately 0.661. This indicates that about 66.1% of the variance in house prices can be explained by the selected features.

```python
# Prepare the data for modeling with the specified features
features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement", "view",
            "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]
X = data[features]
Y = data['price']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X_train, Y_train)

# Make predictions and calculate R^2
Y_pred = model.predict(X_test)
r2 = r2_score(Y_test, Y_pred)

# Print the R^2 value
print('R^2 value:', r2)
```

Question 8
Create a list of tuples, the first element in the tuple contains the name of the estimator:
'scale'
'polynomial'
'model'
The second element in the tuple contains the model constructor
StandardScaler()
PolynomialFeatures(include_bias=False)
LinearRegression()
Use the list to create a pipeline object, predict the 'price', fit the object using the features in the
list features , then fit the model and calculate the R^2

Answer: The pipeline has been successfully created and fitted using the specified features, resulting in an R value of approximately 0.710. This indicates that about 71.0% of the variance in house prices can be explained by the model, which is an improvement over the previous models

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

# Create a list of tuples with estimator names and constructors
estimators = [
    ('scale', StandardScaler()),
    ('polynomial', PolynomialFeatures(include_bias=False)),
    ('model', LinearRegression())
]

# Create a pipeline using the estimators
pipeline = Pipeline(estimators)

# Fit the pipeline to the data
pipeline.fit(X_train[features], Y_train)

# Make predictions and calculate R^2
Y_pred_pipeline = pipeline.predict(X_test[features])
r2_pipeline = r2_score(Y_test, Y_pred_pipeline)

# Print the R^2 value
print('R^2 value from pipeline:', r2_pipeline)
```

Create and fit a Ridge regression object using the training data, setting the regularization

Answer:The Ridge regression model has been successfully fitted with a regularization parameter of 0.1, and the calculated R value is approximately 0.661.
```python
from sklearn.linear_model import Ridge

# Create and fit a Ridge regression model
ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X_train[features], Y_train)

# Make predictions and calculate R^2
Y_pred_ridge = ridge_model.predict(X_test[features])
r2_ridge = r2_score(Y_test, Y_pred_ridge)

# Print the R^2 value
print('R^2 value from Ridge regression:', r2_ridge)
```

Question 10
Perform a second order polynomial transform on both the training data and testing data. Create
and fit a Ridge regression object using the training data, setting the regularization parameter to
0.1. Calculate the R^2 utilizing the test data provided. Take a screenshot of your code and the
R^2.

Answer: The Ridge regression model with a second-order polynomial transformation has been successfully fitted, resulting in an R value of approximately 0.700. This indicates that the model explains about 70.0% of the variance in house prices, which is a solid performance

```python
from sklearn.preprocessing import PolynomialFeatures

# Perform a second order polynomial transform on the training and testing data
poly = PolynomialFeatures(degree=2, include_bias=False)
X_train_poly = poly.fit_transform(X_train[features])
X_test_poly = poly.transform(X_test[features])

# Create and fit a Ridge regression model with the polynomial features
ridge_model_poly = Ridge(alpha=0.1)
ridge_model_poly.fit(X_train_poly, Y_train)

# Make predictions and calculate R^2
Y_pred_ridge_poly = ridge_model_poly.predict(X_test_poly)
r2_ridge_poly = r2_score(Y_test, Y_pred_ridge_poly)

# Print the R^2 value
print('R^2 value from Ridge regression with polynomial features:', r2_ridge_poly)
```
